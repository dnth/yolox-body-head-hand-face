```python
import pandas as pd
import glob

# Define the path to the directory containing the parquet files
# For demonstration purposes, we'll pretend the files are in a directory named 'data/'
# Replace 'data/' with the actual path to your files
directory_path = '../parquets/'
file_pattern = directory_path + 'detection_results_train_chunk_*.parquet'

# Find all parquet files in the directory matching the pattern
parquet_files = glob.glob(file_pattern)

# Read each parquet file and store in a list
dataframes = [pd.read_parquet(file) for file in parquet_files]

# Concatenate all DataFrames into one
combined_dataframe = pd.concat(dataframes, ignore_index=True)

# Save the combined DataFrame to a new parquet file
combined_parquet_path = directory_path + 'detection_results_train.parquet'
combined_dataframe.to_parquet(combined_parquet_path)

# Return the path to the new combined parquet file
combined_parquet_path
```
