```python
import pandas as pd

# Load the large parquet file
df = pd.read_parquet('../parquets/detection_results_train.parquet')
df
```

```python
# Define the chunk size
chunk_size = 1000000  # Adjust based on your needs

# Split and save chunks
for i, chunk in enumerate(range(0, df.shape[0], chunk_size)):
    df_subset = df.iloc[chunk:chunk + chunk_size]
    df_subset.to_parquet(f'../parquets/detection_results_train_chunk_{i}.parquet')
```
